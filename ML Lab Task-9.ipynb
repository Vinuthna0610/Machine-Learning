```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data matrix
data = np.array([(2, 1), (3, 4), (5, 0), (7, 6), (9, 2)])

# Step 1: Compute the mean of each feature
mean = np.mean(data, axis=0)

# Step 2: Center the data by subtracting the mean
centered_data = data - mean

# Step 3: Compute the covariance matrix
cov_matrix = np.cov(centered_data, rowvar=False)

# Step 4: Compute the eigenvalues and eigenvectors of the covariance matrix
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Step 5: Sort the eigenvectors by decreasing eigenvalues
sorted_indices = np.argsort(eigenvalues)[::-1]
sorted_eigenvectors = eigenvectors[:, sorted_indices]

# Step 6: Choose the top k eigenvectors (principal components)
# In this case, we want to transform the data into 2D space, so k=2
k = 2
principal_components = sorted_eigenvectors[:, :k]

# Step 7: Project the centered data onto the principal components to obtain the transformed features
transformed_data = np.dot(centered_data, principal_components)

# Plot the scatter plot of data points in the original domain
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(data[:, 0], data[:, 1], c='blue')
plt.title('Original Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Plot the scatter plot of data points in the transformed domain
plt.subplot(1, 2, 2)
plt.scatter(transformed_data[:, 0], transformed_data[:, 1], c='red')
plt.title('Transformed Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
data = iris.data
target = iris.target

# Step 1: Compute the mean of each feature
mean = np.mean(data, axis=0)

# Step 2: Center the data by subtracting the mean
centered_data = data - mean

# Step 3: Compute the covariance matrix
cov_matrix = np.cov(centered_data, rowvar=False)

# Step 4: Compute the eigenvalues and eigenvectors of the covariance matrix
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Step 5: Sort the eigenvectors by decreasing eigenvalues
sorted_indices = np.argsort(eigenvalues)[::-1]
sorted_eigenvectors = eigenvectors[:, sorted_indices]

# Step 6: Choose the top k eigenvectors (principal components)
# In this case, we want to transform the data into 2D space, so k=2
k = 2
principal_components = sorted_eigenvectors[:, :k]

# Step 7: Project the centered data onto the principal components to obtain the transformed features
transformed_data = np.dot(centered_data, principal_components)

# Plot the scatter plot for samples in the transformed domain with different color codes for samples belonging to different classes
plt.figure(figsize=(8, 6))
for label in np.unique(target):
    plt.scatter(transformed_data[target == label, 0], transformed_data[target == label, 1], label=f'Class {label}')

plt.title('Scatter Plot of Iris Dataset (2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```
